## TIC TAC TOE WITH SELF-ORGANIZING MAPS

Here we decided to create a simple model of self-organizing maps that, trained with points in a 2D plane, is capable of reproduce the same classificiation of a feed-forward NN without the use of the labels.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

We are going to construct a class that will create the network and perform the training. The main important features will be

- **Initialize** the network with number of neurons (classes),number of epochs, batch size, learning rate, radius of the neighborhood
- **Train** by applying the SOM algorithm
- **Predict** once trained

Other features can be useful:

- **Print** the weight configuration
- **Set** a precise weight configuration 



- **Training algorithm**

Once a data is presented to the network, the nearest neuron to the data is calculated (here will be indicated with $u^*$). Next, the weights of all the neurons are updated following the *Kohonen rule*:

Considering a neuron $j$ and a data $\textbf{x}$, the weights $\textbf{w}_{j}$ are update as

$$
\textbf{w}_{j}^{new}= \textbf{w}_{j}^{old} + \eta(t) f(d(j,u^*),\sigma(t))(\textbf{x}-\textbf{w}_{j}^{old})
$$


where $ \eta(t)$ is a step-dependent learning rate, $d(j,u^*)$ is the distance of the $j$-th neuron from the winner neuron and $\sigma(t)$ is the neighborhood radius.


In this case the neghborhood function used is

$$
f(d(j,u^*),\sigma(t)) =e^{-\frac{d^2}{2\sigma^2}}
$$


- **The class**

```python
class SOM(object):
    
    def __init__(self,m,n,num_iterations,batch_size,eta=0.1,sigma=None):
        
        '''
        m x n = dimension of the 2D space in which the map will act, namely the number of neurons of the map
        eta= learning rate
        sigma = radius of the neighborhood
        num_iterations = nuber of training steps
        
        '''
        self._m=m
        self._n=n
        self._num_iterations=num_iterations
        self._eta=float(eta)
        self._batch=batch_size
        
        if sigma is None:
            sigma = max(m,n)/2.0 # Constant radius
            self._sigma=sigma
        else:
            sigma = float(sigma)
            self._sigma = sigma
            
        # initialize the weights randomly
        self._W=tf.random.normal([m*n,2],seed=time.time()) # a matrix of weights 
        self._topography=np.array(list(self._neuron_location(m, n)))
        
        
        #a variable to indicate that the network has been trained
        self._learned=False
        
        
    def train(self,x,i_train):
        # x must be a matrix containing all the points in the dataset
        # i_train is the iteration number
        
        #Create a vector of distances from the data point given
        
        d=[np.sqrt(pow(self._W[i][0]-x[0],2)+pow(self._W[i][1]-x[1],2)) for i in range(len(self._W))]
        
        # Find the smallest d
        
        self._w_idx=np.argmin(d,0)
        self._w_loc=np.array(self._W[self._w_idx])
        
        # Change learning rate and radius as a function of iterations
        learning_rate = 1-i_train/(self._num_iterations*self._batch)
        _eta_new = self._eta * learning_rate
        _sigma_new = self._sigma * learning_rate
        
        # Now update the weights based on how much they are distant from the winner
        # ---> exp(-d**2/2*sigma**2)
        
        d2_neuron=np.array([pow(self._topography[i][0]-self._topography[self._w_idx][0],2)+
                            pow(self._topography[i][1]-self._topography[self._w_idx][1],2) for i in range(len(self._W))])
        
        f_neigh=np.array([np.exp((-1*d2_neuron[i])/(2*pow(_sigma_new,2))) for i in range(len(d2_neuron))])
        
        dW=np.array([_eta_new*f_neigh[i]*(x-self._W[i]) for i in range(len(f_neigh))])
        
        w_new= self._W+dW
        
        self._W=w_new
        
    def fit(self, X):
        count=1
        step=1
        # Train on the dataset num_iteration times
        for i in range(self._num_iterations):
            for x in X:
                self.train(x,count)
                step+=1
                if(step%self._batch==0):
                    count+=1
                
        self._learned=True
    
    def set_network(self,net):
        self._W=net
        #for i in range(len(self._W)):
         #self._W[i]=net[i]
    
    def predict_class(self, X):
        # Given a dataset, it predicts the category of the data by showing the nearest neuron
        if not self._learned:
            raise ValueError("SOM not trained yet")
            
        prediction=[]
        for x in X:
            #Create a vector of distances from the data point given
            d=[np.sqrt(pow(self._W[i][0]-x[0],2)+pow(self._W[i][1]-x[1],2)) for i in range(len(self._W))]
        
            # Find the smallest d
        
            prediction.append(np.argmin(d,0))
            
        prediction=np.array(prediction)
        return prediction
    
    def predict_neuron(self,X):
        #returns a list of the nearest neurons to the dataset examples
        index=self.predict_class(X)
        prediction=[self._topography[index[i]] for i in range(len(index))]
        prediction=np.array(prediction)
        
        return prediction
    
    def print_weights(self):
        #this function return the network
        net=[]
        for i in range(len(self._W)):
            net.append(self._W[i])
            
        net=np.array(net)
        return net
    
    def _neuron_location(self,m,n):
        for i in range(1,m+1):
            for j in range(1,n+1):
                yield np.array([i,j])
                
    def plot_weights(self):
        net=self.print_weights()
        import matplotlib.pyplot as plt
        
        x=[]
        y=[]
        
        for i in range(len(net)):
            x.append(net[i][0])
            y.append(net[i][1])
            
        plt.scatter(x,y,marker='o')
        plt.xlabel("x")
        plt.ylabel("y")
        plt.title("Positions of the neuron weights")
        
        plt.show()
        plt.close()
```

- **Preparing the dataset**

Now that we have the class to create a self-organizing map, we have to import the dataset

```python
n_train = 500
n_valid = 100
# generate synthetic training dataset
x_train = np.zeros((n_train, 2))
y_train = np.zeros((n_train, 9))
# generate synthetic validation data
x_valid = np.zeros((n_valid, 2))
y_valid = np.zeros((n_valid, 9))

sigma = 0.2

for i in range(n_train):
    j = np.random.randint(0,9)
    y_train[i,j]=1
    if j == 0:
        mux, muy = 1, 1
    elif j == 1:
        mux, muy = 2, 1
    elif j == 2:
        mux, muy = 3, 1
    elif j == 3:
        mux, muy = 1, 2
    elif j == 4:
        mux, muy = 2, 2
    elif j == 5:
        mux, muy = 3, 2
    elif j == 6:
        mux, muy = 1, 3
    elif j == 7:
        mux, muy = 2, 3
    else:
        mux, muy = 3, 3
    x_train[i,0] = np.random.normal(mux, sigma)
    x_train[i,1] = np.random.normal(muy, sigma)

for i in range(n_valid):
    j = np.random.randint(0,9)
    y_valid[i,j]=1
    if j == 0:
        mux, muy = 1, 1
    elif j == 1:
        mux, muy = 2, 1
    elif j == 2:
        mux, muy = 3, 1
    elif j == 3:
        mux, muy = 1, 2
    elif j == 4:
        mux, muy = 2, 2
    elif j == 5:
        mux, muy = 3, 2
    elif j == 6:
        mux, muy = 1, 3
    elif j == 7:
        mux, muy = 2, 3
    else:
        mux, muy = 3, 3
    x_valid[i,0] = np.random.normal(mux, sigma)
    x_valid[i,1] = np.random.normal(muy, sigma)

plt.plot(x_train[:,0],x_train[:,1],'o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

In this case we won't need the <code>y_train</code>, maybe <code>y_valid</code> for a check

- **Create the network and train**

First we create a network and print the weights that are randomly initialized. Over 500 data we can make a single training epoch and batches of size 30 (with batches here we indicate the amount of examples after which the radius and the learning rate are shrinked).

```python
# Create a network with 9 classes, that makes a single training step over the dataset
model=SOM(3,3,1,30)
```

```python
model.plot_weights()
```

Then the network is trained and the new positions of the weights are plotted.

```python
model.fit(x_train)
```

```python
model.plot_weights()
```

- **Predicting and classifying the data**

We can now use the network for classifying the <code>x_valid</code> set.

```python
pred=model.predict_class(x_valid)
```

Now that we have the prediction, apply the label to each <code> x_valid</code> data and plot them.

```python
plt.scatter(x_valid[:,0], x_valid[:,1], c=pred)

bounds = np.linspace(0,max(pred),max(pred)+1)
cb = plt.colorbar(ticks=bounds)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Class prediction from self organizing map")

plt.show()
```

As we can see, the self organizing map performs quite well in classifying the validation data.
